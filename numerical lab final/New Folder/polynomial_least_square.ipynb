```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data points
x = np.array([1, 2, 3, 4, 5], dtype=int)
y = np.array([3.5, 6.2, 10.8, 17.3, 25.6], dtype=float)

# Degree of the polynomial to fit
n = 2
m = len(x)

# step-1
x_sum = [np.sum(x**k) for k in range(2 * n + 1)]
xy_sum = [np.sum((x**k)*y) for k in range(n+1)]

# step-2
A = np.zeros((n+1, n+1))
B = np.zeros(n+1)

# step-3
for i in range(n+1):
    for j in range(n+1):
        A[i, j] = x_sum[i+j]
    B[i] = xy_sum[i]


a = np.linalg.solve(A, B)
print("Polynomial coefficients (a0, a1, a2) : ", a)

# step-4

y_pred = np.zeros_like(x, dtype=float)
for i in range(n + 1):
    y_pred += a[i] * (x ** i)


# plot

plt.scatter(x, y, color="red", label="Data Points")
plt.plot(x, y_pred, color="darkblue", label=f"Degree{n} fit")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.title("Polynomial Least Squares Fit")
plt.show()
```
