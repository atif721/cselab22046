```python
import math as MT


def func(x):
    return 1 / (1 + x**2)


def simpson_rule(lower, upper, intervals):
    step = (upper - lower) / intervals
    summation = func(lower) + func(upper)
    for idx in range(1, int(intervals)):
        if idx % 2 == 0:
            summation += 2 * func(lower + idx * step)
        else:
            summation += 4 * func(lower + idx * step)
    return (step / 3) * summation


lower_limit = float(input("Enter lower limit (Xo): "))
upper_limit = float(input("Enter upper limit (Xn): "))
num_intervals = int(input("Enter number of intervals (n, even number): "))

if num_intervals % 2 != 0:
    print("Number of intervals (n) must be even for Simpson's rule.")
else:
    print("=> By using Simpson rule:")
    approximation = simpson_rule(lower_limit, upper_limit, num_intervals)
    exact_value = MT.atan(upper_limit) - MT.atan(lower_limit)
    absolute_error = abs(exact_value - approximation)

    print(f"For limit ({lower_limit}, {upper_limit})", end=" ")
    print(f"and interval: {num_intervals}\n")

    print(f"Approximate value of the integral: {approximation:.5f}")
    print(f"Exact value of the integral:        {exact_value:.5f}")
    print(f"Absolute error:                     {absolute_error:.5f}")
```
